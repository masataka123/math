```python
from sympy import *
import numpy as np
```

```python
A=symbols('a:9')
A=np.array(A).reshape(9,1)
B=symbols('b:21')
B=np.array(B).reshape(7,3)
C=symbols('c:25')
C=np.array(C).reshape(5,5)
D=symbols('d:21')
D=np.array(D).reshape(3,7)
E=symbols('e:9')
E=np.array(E).reshape(1,9)
```

```python
x=Symbol('x')
y=Symbol('y')
s=Symbol('s')
t=Symbol('t')
```

```python
def F(x):
    poly=0
    for i in range(9):
        poly+=A[i][0]*(x**i)
    return poly
```

```python
def G(x,y):
    poly=0
    for i in range(7):
        for j in range(3):
            poly+=B[i][j]*(x**i)*(y**j)
    return poly
```

```python
def H(x,y):
    poly=0
    for i in range(5):
        for j in range(5):
            poly+=C[i][j]*(x**i)*(y**j)
    return poly
```

```python
def I(x,y):
    poly=0
    for i in range(3):
        for j in range(7):
            poly+=D[i][j]*(x**i)*(y**j)
    return poly
```

```python
def J(y):
    poly=0
    for i in range(9):
        poly+=E[0][i]*(y**i)
    return poly
```

```python
def rela(s,t,alpha,beta):
    X=s+alpha
    Y=s*t+beta
    L1=F(X)*t+G(X,Y)
    L2=F(X)*t*t+G(X,Y)*t+H(X,Y)
    L3=F(X)*t*t*t+G(X,Y)*t*t+H(X,Y)*t+I(X,Y)
    L4=F(X)*(t**4)+G(X,Y)*(t**3)+H(X,Y)*(t**2)+I(X,Y)*t+J(Y)
    f1=div(L1,s,s)[1]
    f2=div(L2,s**2,s)[1]
    f3=div(L3,s**3,s)[1]
    f4=div(L4,s**4,s)[1]
    return f1,f2,f3,f4
```

```python
blow0=rela(s,t,0,0)
blow0
```

```python
linear=[]
blow0=rela(s,t,0,0)
for k in range(4):
    po=Poly(blow0[k],s,t)
    linear+=po.coeffs()
    
blow1=rela(s,t,1,1)
for k in range(4):
    po=Poly(blow1[k],s,t)
    linear+=po.coeffs()
    
blow2=rela(s,t,-1,-1)
for k in range(4):
    po=Poly(blow2[k],s,t)
    linear+=po.coeffs()
    
blow3=rela(s,t,2,3)
for k in range(4):
    po=Poly(blow3[k],s,t)
    linear+=po.coeffs()
```
